```python
import zipfile

input_dir = r"F:\23 PARTIFLUX\input_FLUXNET2025"
output_dir = r"F:\23 PARTIFLUX\HH_FLUXNET2025"

os.makedirs(output_dir, exist_ok=True)

# Loop through all zip files
for fname in os.listdir(input_dir):
    if fname.endswith(".zip"):
        zip_path = os.path.join(input_dir, fname)
        
        try:
            with zipfile.ZipFile(zip_path, 'r') as z:
                for file in z.namelist():
                    if file.endswith('.csv') and 'HH' in os.path.basename(file):
                        out_path = os.path.join(output_dir, os.path.basename(file))
                        
                        with z.open(file) as source, open(out_path, 'wb') as target:
                            target.write(source.read())
                        
                        # Print the filename that was extracted
                        print(f"Extracted: {os.path.basename(file)}")
        
        except zipfile.BadZipFile:
            print(f"Warning: '{zip_path}' is not a valid zip file and was skipped.")

print("Done extracting all HH CSV files.")
```
